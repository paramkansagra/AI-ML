Hierarchical Clustering 

It is similar to KMeans clustering like we have normal data at first and then we would get clustered data after applying clustering

There are 2 types under hierarchical clustering i.e. agglomerative(bottom up approch) and divisive(top to bottom approch)

Agglomerative Approch:-
    1. Make each data point to a single point cluster i.e. form N clusters
    2. Then take 2 closest points and make the cluster i.e. form N-1 clusters
    3. Then take those 2 closest clusters and make then one cluster i.e. form N-2 clusters
    4. Repeat step 3 until there is only 1 cluster left.

The distance between the clusters is found out by the Euclidian distance.

The closeness of clusters could be determined by:-
    1. We can find the distance between 2 clusters by the closest points
    2. We can find the distance between 2 clusters by the farthest points
    3. Take the average distance of points
    4. Find the centroids of the clusters and then find the distance

Dendograms

Import the modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import the dataset

```python
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:,[3,4]].values
```

Using the dendogram to find the number of optimal clusters

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method = "ward"))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Eucledian distances")
plt.show()
```

Training the hirearchial model

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5 , affinity = "euclidean" , linkage = "ward")
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```

Visualize the Clustering 

```python
# we would be using multiple scatter plots to get the clusters

plt.scatter(X[y_hc == 0,0] , X[y_hc == 0 , 1] , s = 50 ,c="red", label = "cluster 1")
plt.scatter(X[y_hc == 1,0] , X[y_hc == 1 , 1] , s = 50 , c="blue", label = "cluster 2")
plt.scatter(X[y_hc == 2,0] , X[y_hc == 2 , 1] , s = 50 , c="green", label = "cluster 3")
plt.scatter(X[y_hc == 3,0] , X[y_hc == 3 , 1] , s = 50 , c="cyan", label = "cluster 4")
plt.scatter(X[y_hc == 4,0] , X[y_hc == 4 , 1] , s = 50 , c="magenta", label = "cluster 5")
plt.title("Cluster of customers")
plt.xlabel("Annual income")
plt.ylabel("Spending score")
plt.legend()
plt.show()
```
