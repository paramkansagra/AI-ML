Import the modules

```python
# import the libraries for using

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Make the X and Y axis and open the file

```python
# import the dataset and then split the data into X and Y axis

dataset = pd.read_csv("Salary_Data.csv")
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
```

Splitting the data into test and train set

```python
# split the data into a train and test set

from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

Training the data on simple regression model

```python
# make a linear model and import the linear regression model

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,Y_train)

# now we are fitting the train result into the X and Y train 
# after fitting we will get a regressor model
```

Predict the Test Results

```python
# we are predicting the Y values using the regressor model and pushing it into the Y_predicted values

Y_predicted = regressor.predict(X_test)
```

Visualizing Training set results

```python
# visualizing the training set results
# the X axis will be the experince
# the Y axis will be the salary

# red point ==> real salary
# blue point ==> predicted salary

plt.scatter(X_train,Y_train,color = 'red')

# plotting the regression line
plt.plot(X_train,regressor.predict(X_train),color = 'blue')

plt.title("Salary v/s exprince (Training set)")
plt.xlabel("Years of Experince")
plt.ylabel("Salary")

plt.show() # to show the graph
```

Visualizing the Test set results

```python
plt.scatter(X_test,Y_test,color="red") #plotting the the test set points

plt.plot(X_test,Y_predicted,color = "blue") #plotting the regressor line

plt.title("Predicted salary v/s years of experince") # making the title

plt.xlabel("Years Of experince") # x axis label

plt.ylabel("Salary") # y axis label

plt.show()
```
