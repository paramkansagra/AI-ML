Import all the nessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import the dataset 

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
Y = dataset.iloc[:,-1].values
```

Train the Linear regression model on Training set

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X,Y)
```

Train the polynomial regression model on Training set

```python
# polynomial regressor is nothing but y = b1 + b2 * x + b3 * x^2 + b4 * x^3 and so on

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 2) # we are squaring the values at 2

X_poly = poly_reg.fit_transform(X)

polynomial_regressor = LinearRegression()
polynomial_regressor.fit(X_poly,Y)
```

Visualizing the linear Regressor models

```python
plt.scatter(X,Y,color = 'red')
plt.plot(X,linear_regressor.predict(X),color = "blue")
plt.title("Linear regressor model")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

Visualizing Polynomial Regressor Model

```python
plt.scatter(X,Y,color="red")
plt.plot(X,polynomial_regressor.predict(X_poly),color="blue")
plt.title("Polynomial Regressor")
plt.xlabel("Level squared")
plt.ylabel("Salary")
plt.show()
```

Predict a result with linear regression

```python
linear_regressor.predict([[6.5]]) # a person with level 6.5 is having the salary as output meaning it is incorrect
```

Predict a result with multiple regression

```python
polynomial_regressor.predict(poly_reg.transform([[6.5]]))

# here the output comes out to be this which is much more fitting to the data we have
```
